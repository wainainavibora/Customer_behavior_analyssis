```python
#age _groups
labels = ['young_adult', 'adult','Middle-aged', 'senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
frequency_mapping = {
    'fortnightly': 14,
    'weekly': 7,
    'monthly': 30,
    'quarterly': 90,
    'bi-weekly': 14,
    'annually': 365,
    'every_3_months': 90,
}
```

```python
frequency_mapping = {k.lower(): v for k, v in frequency_mapping.items()}
if 'forthnightly' in frequency_mapping:
    frequency_mapping['fortnightly'] = frequency_mapping.pop('forthnightly')

df['purchase_frequency_days'] = df['frequency_of_purchases'].str.strip().str.lower().map(frequency_mapping)
df['purchase_frequency_days'] = df['purchase_frequency_days'].fillna(df['purchase_frequency_days'].median())

df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```
